## Importando bibliotecas

```python
import pandas as pd
import numpy as np
```

# Tratamento de dados de hospedagem

```python
dados = pd.read_json('/content/moveis_disponiveis.json')
dados.head()
```

```python
# Normalizando objeto e convertendo para DataFrame
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
```

## Método explode
Transforma campos que são listas em linhas individuais no DataFrame

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
dados.head()
```

## Método reset index
Altera índices do DataFrame de forma a deixa-los ordenados

```python
# Inplace = True: substitui diretamente no DataFrame, drop = True: não mantêm índices antigos
dados.reset_index(inplace = True, drop= True)
dados.head()
```

## Método Info
Resume os dados do DataFrame

```python
dados.info()
```

## Método Astype
Converte dado para o tipo desejado

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
# Convertendo colunas simultaneamente
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
dados.info()
```

## Tratando colunas com caracteres indesejados

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',','').strip())
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

## Tratando dados textuais

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
# Removendo caracteres indesejados com regex (excesão)
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex = True)
```

```python
# Exclui hífens que estão fora das palavras (travessão)
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

```python
# Tokenizando dados
dados['descricao_local'] = dados['descricao_local'].str.split()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
```

```python
dados.head()
```

# Tratamento de dados de vendas

```python
dados_clientes = pd.read_json('/content/dados_vendas_clientes.json')
dados_clientes.head()
```

```python
# Normalizando dados
dados_clientes = pd.json_normalize(dados_clientes['dados_vendas'])
dados_clientes.head()
```

```python
colunas = list(dados_clientes)
dados_clientes = dados_clientes.explode(colunas[1:])
```

```python
dados_clientes.reset_index(inplace = True, drop = True)
```

```python
dados_clientes.head()
```

```python
# Tratando textos
dados_clientes['Valor da compra'] = dados_clientes['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',','').strip())
dados_clientes['Valor da compra'] = dados_clientes['Valor da compra'].astype(np.float64)
```

```python
dados_clientes['Cliente'] = dados_clientes['Cliente'].str.lower()
dados_clientes['Cliente'] = dados_clientes['Cliente'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex = True)
dados_clientes['Cliente'] = dados_clientes['Cliente'].str.replace('[0-9]', ' ', regex = True)
```

```python
# Convertendo para datetime
dados_clientes['Data de venda'] = pd.to_datetime(dados_clientes['Data de venda'])
```

```python
dados_clientes.info()
```

```python
dados_clientes.head()
```

# Tratamento de dados de condomínio

```python
dados_imoveis = pd.read_json('/content/dados_locacao_imoveis.json')
dados_imoveis.head()
```

```python
dados_imoveis = pd.json_normalize(dados_imoveis['dados_locacao'])
```

```python
dados_imoveis.head()
```

```python
colunas = list(dados_imoveis.columns)
colunas

dados_imoveis = dados_imoveis.explode(colunas[1:])
```

```python
dados_imoveis.reset_index(inplace = True, drop = True)
```

```python
dados_imoveis['valor_aluguel'] = dados_imoveis['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',','.').replace('reais','').strip())
dados_imoveis['valor_aluguel'] = dados_imoveis['valor_aluguel'].astype(np.float64)
```

```python
dados_imoveis['datas_combinadas_pagamento'] = pd.to_datetime(dados_imoveis['datas_combinadas_pagamento'], dayfirst=True)
dados_imoveis['datas_de_pagamento'] = pd.to_datetime(dados_imoveis['datas_de_pagamento'], dayfirst=True)
```
